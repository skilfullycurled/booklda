```python
files = [filename.split('.')[0] for filename in os.listdir('digitalSTS/papers/dicts/') if not filename.startswith('.')]
```

```python
filename = files[0]
dict_path = 'digitalSTS/papers/vis/dicts/'
corpus_path = 'digitalSTS/papers/vis/corpuses/'
lda_path = 'digitalSTS/papers/vis/lda/'

dict_filename   = dict_path + filename + '.dict'
corpus_filename = corpus_path + filename + '.mm'
lda_filename    = lda_path + filename + '.lda'

dictionary = corpora.Dictionary.load(dict_filename)
corpus = corpora.MmCorpus(corpus_filename)
lda = models.LdaModel.load(lda_filename)

essay =  pyLDAvis.gensim.prepare(lda, corpus, dictionary)
print filename
pyLDAvis.display(essay)
```
